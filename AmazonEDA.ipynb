```python
df=df.dropna()
```

```python
df.rating=df.rating.replace(["|"],["4.5"])
```

```python
df["rating"]=df[["rating"]].astype(float)
```

```python
# What is the average rating for each product category
df.groupby("category")["rating"].mean()
```

```python
#2 What are the top rating_count products by category
df.sort_values("rating_count").groupby("category")["rating_count"].sum()[0:5]
```

```python
df["actual_price"]=df["actual_price"].str.replace(',','')
```

```python
df["discounted_price"]=df["discounted_price"].str.replace(',','')
```

```python
df["actual_price"]=df["actual_price"].str.replace('₹','')
```

```python
df["discounted_price"]=df["discounted_price"].str.replace('₹','')
```

```python
df["actual_price"]=df["actual_price"].astype(float)
```

```python
df["discounted_price"]=df["discounted_price"].astype(float)
```

```python
#P What is the distribution of discounted prices vs. actual prices
df['actual_price'].hist(bins=50)
plt.xlabel('Price')
plt.title('Distribution of actual prices')
plt.ylabel('Price')
plt.show()
df['discounted_price'].hist(bins=50)
plt.xlabel('Price')
plt.title('Distribution of discounted prices')
plt.ylabel('Price')
plt.show()
```
